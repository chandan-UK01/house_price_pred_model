```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,10)
```

```python
df1=pd.read_csv("Bengaluru_House_Data.csv")
df1.head()
```

```python
df1.shape
```

```python
df1.groupby("area_type")["area_type"].agg("count")
```

```python
df2=df1.drop(["area_type","society","balcony","availability"],axis=1)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3=df2.dropna()
df3.isnull().sum()
```

```python
df3.shape
```

```python
df3["size"].unique()
```

```python
df3["bhk"]=df3["size"].apply(lambda x : int(x.split(" ")[0]))
df3.head()
```

```python
df3["bhk"].unique()
```

```python
df3[df3.bhk>20]
```

```python
df3.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3["total_sqft"].apply(is_float)].head(10)
```

```python
def total_sqft_to_single_num(x):
    tokens=x.split("-")
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4=df3.copy()
df4["total_sqft"]=df4["total_sqft"].apply(total_sqft_to_single_num)
df4.head()
```

```python
df5=df4.copy()
df5["price_per_sqft"]=df5["price"]*100000/df5["total_sqft"]
df5.head()
```

```python
len(df5["location"].unique())
```

```python
df5.location=df5.location.apply(lambda x: x.strip())

location_statistics=df5.groupby("location")["location"].agg("count").sort_values(ascending=False)
location_statistics
```

```python
len(location_statistics[location_statistics<=10])
```

```python
len(df5.location.unique())
```

```python
df5.location=df5.location.apply(lambda x : "others" if x in location_statistics[location_statistics<=12] else x)
```

```python
len(df5.location.unique())
```

```python
df5.head(10)
```

```python
df5[df5.total_sqft/df5.bhk<300].head()
```

```python
df5.shape
```

```python
df6=df5[~(df5.total_sqft/df5.bhk<300)]
df6.shape
```

```python
df6.price_per_sqft.describe()
```

```python
def remove_outliers(df):
    df_out=pd.DataFrame()
    for key, subdf in df.groupby("location"):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df=subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft <=(m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out

df7=remove_outliers(df6)

df7.shape
```

```python
df7.bath.unique()
```

```python
df7[df7.bath>10]
```

```python
df7[df7.bath>df7.bhk+2]
```

```python
df8=df7[df7.bath<df7.bhk+2]
df8.shape
```

```python
df9=df8.drop(["size","price_per_sqft"],axis=1)
df9.head()
```

```python
dummies=pd.get_dummies(df9.location)
dummies.head(20)
```

```python
df10=pd.concat([df9,dummies.drop("others",axis=1)],axis=1)
df10.head()
```

```python
df11=df10.drop("location",axis=1)
df11.head(10)
```

```python
df11.shape
```

```python
X=df11.drop("price",axis=1)
X.head()
```

```python
y=df11.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_trian,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_trian)
lr.score(X_test,y_test)
```

```python
X.columns
```

```python
def predict_price(location,sqf,bath,bhk):
    loc_index=np.where(X.columns==location)[0][0]
    
    x=np.zeros(len(X.columns))
    x[0]=sqf
    x[1]=bath
    x[2]=bhk
    if loc_index >=0:
        x[loc_index]=1
        
    return lr.predict([x])[0]
```

```python
predict_price("1st Phase JP Nagar",1000,2,3)
```
